To repair the teleporter, you first need to understand the beam-splitting properties of the tachyon manifold. In this example, a tachyon beam is split a total of 21 times.

Analyze your manifold diagram. How many times will the beam be split?

```python
with open('test_input.txt', 'r') as f: 
    test_rows = []
    for line in f: 
        test_rows.append(line.strip())

test_rows
```

```python
def get_beam_split_count(rows):
    def find_char_in_string(char, string): 
        return {i for i, x in enumerate(string) if x == char}

    split_counter = 0

    beam_indices = find_char_in_string('S', rows[0])
    print('b', beam_indices)

    for row in rows[1:]: 
        splitter_indices = find_char_in_string('^', row)
        if not splitter_indices: 
            continue
        print('s', splitter_indices)

        splitting_locs = beam_indices.intersection(splitter_indices)
        print('matches', splitting_locs)
        split_counter += len(splitting_locs)

        for i in splitting_locs: 
            beam_indices.update({i-1, i+1})
            beam_indices.remove(i)

    return split_counter
```

With your analysis of the manifold complete, you begin fixing the teleporter. However, as you open the side of the teleporter to replace the broken manifold, you are surprised to discover that it isn't a classical tachyon manifold - it's a quantum tachyon manifold.

To fix the manifold, what you really need to know is the number of timelines active after a single particle completes all of its possible journeys through the manifold.

In this example, in total, the particle ends up on 40 different timelines.

Apply the many-worlds interpretation of quantum tachyon splitting to your manifold diagram. In total, how many different timelines would a single tachyon particle end up on?



```python
def get_beam_split_timeline_count(rows):
    def find_char_in_string(char, string): 
        return {i for i, x in enumerate(string) if x == char}

    split_counter = 0
    beam_counts = dict()
    for i in range(len(rows[0])): 
        beam_counts[i] = 0
        
    beam_indices = find_char_in_string('S', rows[0])
    # print('b', beam_indices)
    
    for i in beam_indices: 
        beam_counts[i] = beam_counts[i] + 1

    for row in rows[1:]: 
        splitter_indices = find_char_in_string('^', row)
        if not splitter_indices: 
            continue
        # print('s', splitter_indices)

        splitting_locs = beam_indices.intersection(splitter_indices)
        # print('matches', list(splitting_locs))
        split_counter += len(splitting_locs)

        for i in splitting_locs: 
            beam_counts[i-1] = beam_counts[i-1] + beam_counts[i]
            beam_counts[i+1] = beam_counts[i+1] + beam_counts[i]
            beam_counts[i] = 0

            beam_indices.update({i-1, i+1})
            beam_indices.remove(i)

        # print(beam_counts)
        # print('\n ------ \n')

    # print(sum(beam_counts.values()))

    return sum(beam_counts.values())
```
