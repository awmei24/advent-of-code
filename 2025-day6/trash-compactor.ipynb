As you try to find a way out, you are approached by a family of cephalopods! They're pretty sure they can get the door open, but it will take some time. While you wait, they're curious if you can help the youngest cephalopod with her math homework.

Cephalopod math doesn't look that different from normal math. The math worksheet (your puzzle input) consists of a list of problems; each problem has a group of numbers that need to be either added (+) or multiplied (*) together.

**_However, the problems are arranged a little strangely; they seem to be presented next to each other in a very long horizontal list. For example_**:

- 123 328  51 64 
-  45 64  387 23 
-   6 98  215 314
- x +   x   +

Each problem's numbers are arranged vertically; at the bottom of the problem is the symbol for the operation that needs to be performed. Problems are separated by a full column of only spaces. The left/right alignment of numbers within each problem can be ignored.

**_So, this worksheet contains four problems_**:

- 123 * 45 * 6 = 33210
- 328 + 64 + 98 = 490
- 51 * 387 * 215 = 4243455
- 64 + 23 + 314 = 401

To check their work, cephalopod students are given the grand total of adding together all of the answers to the individual problems. In this worksheet, the grand total is 33210 + 490 + 4243455 + 401 = 4277556.

Of course, the actual worksheet is much wider. You'll need to make sure to unroll it completely so that you can read the problems clearly.

Solve the problems on the math worksheet. What is the grand total found by adding together all of the answers to the individual problems?

```python
with open('test_input.txt', 'r') as f: 
    test_rows = []
    for line in f: 
        test_rows.append(line.strip())
    
    updated_test_rows = []
    for test_row in test_rows: 
        updated_test_row = []
        split_test_row = test_row.split(' ')
        for element in split_test_row: 
            if element: 
                updated_test_row.append(element)

        updated_test_rows.append(updated_test_row)

updated_test_rows
```

```python
import pandas as pd

def get_grand_total(updated_rows):
    def get_sum(values): 
        out = 0
        for value in values: 
            out += int(value)
        
        return out

    def get_product(values): 
        out = 1
        for value in values: 
            out *= int(value)
        
        return out

    col_list = []
    for i in range(len(updated_rows[0])): 
        col_list.append(f'P{i}')

    df = pd.DataFrame(updated_rows, columns=col_list)

    final_sum = 0
    for problem_col in df: 
        problem = df[problem_col].tolist()

        if problem[-1] == '+': 
            answer = get_sum(problem[:-1])
        
        if problem[-1] == '*': 
            answer = get_product(problem[:-1])
        
        final_sum += answer

    return final_sum
```

```python
get_grand_total(updated_test_rows)
```

```python
with open('input.txt', 'r') as f: 
    rows = []
    for line in f: 
        rows.append(line.strip())
    
    updated_rows = []
    for row in rows: 
        updated_row = []
        split_row = row.split(' ')
        for element in split_row: 
            if element: 
                updated_row.append(element)

        updated_rows.append(updated_row)

get_grand_total(updated_rows)
```

(Problems are still separated with a column consisting only of spaces, and the symbol at the bottom of the problem is still the operator to use.)

- 123 328  51 64 
-  45 64  387 23

The third problem from the right is 8 + 248 + 369 = 625

Now, the grand total is 1058 + 3253600 + 625 + 8544 = 3263827.
