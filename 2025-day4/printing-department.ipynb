You ride the escalator down to the printing department. They're clearly getting ready for Christmas; they have lots of large rolls of paper everywhere, and there's even a massive printer in the corner (to handle the really big print jobs).

The forklifts can only access a roll of paper if there are fewer than four rolls of paper in the eight adjacent positions. If you can figure out which rolls of paper the forklifts can access, they'll spend less time looking and more time breaking down the wall to the cafeteria.

**_In this example, there are 13 rolls of paper that can be accessed by a forklift (marked with x)_**:

Consider your complete diagram of the paper roll locations. How many rolls of paper can be accessed by a forklift?

```python
import itertools

def get_accessible_rolls(rows):
    def count_rolls_in_surroundings(locations, x, y): 
        all_surrounding = list(itertools.product((x - 1, x, x + 1), (y - 1, y, y + 1)))
        
        roll_count = 0
        
        for i, j in all_surrounding: 
            if (i, j) == (x, y): 
                continue

            if locations[i][j] == '@': 
                roll_count += 1
            
        return roll_count

    total_rolls = 0

    total_rows = len(rows)
    total_cols = len(rows[0])

    padding_row = ['.' * (total_cols + 2)]
    padded_rows = padding_row + ['.' + row + '.' for row in rows] + padding_row

    for ix in range(1, total_rows + 1): 
        for iy in range(1, total_cols + 1): 
            if padded_rows[ix][iy] != '@': 
                continue

            roll_count = count_rolls_in_surroundings(padded_rows, ix, iy)

            if roll_count < 4: 
                total_rolls += 1

    return total_rolls
```

Stop once no more rolls of paper are accessible by a forklift. In this example, a total of 43 rolls of paper can be removed.

Start with your original diagram. How many rolls of paper in total can be removed by the Elves and their forklifts?

```python
import itertools

def pad_rows(rows): 
    total_cols = len(rows[0])

    padding_row = ['.' * (total_cols + 2)]
    padded_rows = padding_row + ['.' + row + '.' for row in rows] + padding_row

    return padded_rows

def get_accessible_rolls_and_locs(padded_rows):
    def count_rolls_in_surroundings(locations, x, y): 
        all_surrounding = list(itertools.product((x - 1, x, x + 1), (y - 1, y, y + 1)))
        
        roll_count = 0
        
        for i, j in all_surrounding: 
            if (i, j) == (x, y): 
                continue

            if locations[i][j] == '@': 
                roll_count += 1
            
        return roll_count

    total_rolls = 0
    accessible_locations = []

    total_rows = len(padded_rows) - 2
    total_cols = len(padded_rows[0]) - 2

    for ix in range(1, total_rows + 1): 
        for iy in range(1, total_cols + 1): 
            if padded_rows[ix][iy] != '@': 
                continue

            roll_count = count_rolls_in_surroundings(padded_rows, ix, iy)

            if roll_count < 4: 
                total_rolls += 1
                accessible_locations.append((ix, iy))

    return total_rolls, accessible_locations

def remove_rolls(padded_rows, accessible_locations): 
    new_padded_rows = []
    for ix in range(len(padded_rows)): 
        new_row = []
        for iy in range(len(padded_rows[0])):
            # print(ix, iy)
            # print(padded_rows[ix][iy])
            if (ix, iy) in accessible_locations: 
                new_row.append('.')
            else: 
                new_row.append(padded_rows[ix][iy])
        new_padded_rows.append(new_row)
    
    return new_padded_rows

def get_all_accessible_rolls(rows): 
    padded_rows = pad_rows(rows)
    none_removable = False
    total_rolls = 0

    while not none_removable: 
        rolls_n, accessible_locations = get_accessible_rolls_and_locs(padded_rows)
        # print(rolls_n)
        # print(accessible_locations)

        total_rolls += rolls_n
        padded_rows = remove_rolls(padded_rows, accessible_locations)

        # for row in padded_rows: 
        #     print(row)
        
        # print('----------')

        if rolls_n == 0: 
            none_removable = True

    return total_rolls
```
