Using two red tiles as opposite corners, what is the largest area of any rectangle you can make?

```python
import math
import itertools

def get_largest_rectangle(raw_coordinates):

    def get_area(coord_1, coord_2): 
        return (abs(coord_2[1] - coord_1[1]) + 1) * (abs(coord_2[0] - coord_1[0]) + 1)
    
    coordinates = []
    for coordinate in raw_coordinates: 
        x, y = map(int, coordinate.split(','))
        coordinates.append((x, y))

    pairwise_combinations = list(itertools.combinations(coordinates, 2))

    all_areas = []
    for pair in pairwise_combinations: 
        all_areas.append((pair[0], pair[1], get_area(pair[0], pair[1])))

    all_areas.sort(key=lambda x: x[2], reverse=True)
    
    return all_areas[0][2]
```

Using two red tiles as opposite corners, what is the largest area of any rectangle you can make using only red and green tiles?

```python
from shapely.geometry import Polygon

def get_largest_green_rectangle(raw_coordinates): 
    def get_area(coord_1, coord_2): 
        return (abs(coord_2[1] - coord_1[1]) + 1) * (abs(coord_2[0] - coord_1[0]) + 1)

    coordinates = []
    for coordinate in raw_coordinates: 
        x, y = map(int, coordinate.split(','))
        coordinates.append((x, y))

    # get areas
    pairwise_combinations = list(itertools.combinations(coordinates, 2))

    all_areas = []
    for pair in pairwise_combinations: 
        coord_1, coord_2 = pair
        x1, y1 = coord_1
        x2, y2 = coord_2

        all_areas.append(([(x1, y1), (x2, y1), (x2, y2), (x1, y2)], get_area(pair[0], pair[1])))

    all_areas.sort(key=lambda x: x[1], reverse=True)

    valid_area = Polygon(coordinates)

    for area in all_areas:
        rectangle = Polygon(area[0])
        
        if valid_area.contains(rectangle):
            return area[1] 
    
    return
```

```python
with open('test_input.txt', 'r') as f: 
    test_raw_coordinates = []
    for line in f: 
        test_raw_coordinates.append(line.strip())

get_largest_green_rectangle(test_raw_coordinates)
```

```python
with open('input.txt', 'r') as f: 
    raw_coordinates = []
    for line in f: 
        raw_coordinates.append(line.strip())

get_largest_green_rectangle(raw_coordinates)
```
